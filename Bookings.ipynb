```python
# Задача найти Churn Rate 
```

```python
arriva_free_baby = bookings.query('total_kids == 0 and is_canceled == 0')
# приехали без детей и НЕ отменили бронь
```

```python
nearriva_free_baby= bookings.query('total_kids == 0 and is_canceled == 1')
# приехали без детей и ОТМЕНИЛИ бронь
```

```python
total_free_baby = arriva_free_baby.shape[0] + nearriva_free_baby.shape[0]
# общее количество клиентов без детей
```

```python
# приехали с детьми и НЕ отменили бронь
arriva_with_baby= bookings.query('total_kids > 0 and is_canceled == 0')
```

```python
# приехали с детьми и ОТМЕНИЛИ бронь
nearriva_with_baby= bookings.query('total_kids > 0 and is_canceled == 1')
```

```python
# общее количество клиентов с детьми
total_with_baby= arriva_with_baby.shape[0] + nearriva_with_baby.shape[0]
```

```python
# Получение Churn Rate для клиентов без детей
ChurnRate_free_baby = (nearriva_free_baby.shape[0] / total_free_baby) * 100
```

```python
ChurnRate_free_baby_result = round(ChurnRate_free_baby, 2)
```

```python
ChurnRate_free_baby_result
```

```python
# Получение Churn Rate для клиентов с детьми
ChurnRate_with_baby= (nearriva_with_baby.shape[0] / total_with_baby) * 100
```

```python
ChurnRate_with_baby_result= round(ChurnRate_with_baby, 2)
```

```python
ChurnRate_with_baby_result
```

```python
max_churn_rate = max(ChurnRate_with_baby_result, ChurnRate_free_baby_result)
```
